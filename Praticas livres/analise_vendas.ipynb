
# Análise de Vendas
# Dataset de exemplo com vendas mensais.


# Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Criar DataFrame de Exemplo

# Gerar dados fictícios

```python
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = np.random.randint(100, 500, size=12)
products = ['Produto A', 'Produto B', 'Produto C'] * 4

df = pd.DataFrame({
    'Data': dates,
    'Produto': products,
    'Vendas': sales
})
```

```python
# Exibir primeiras linhas
# %%
# Pivotar a tabela para separar produtos em colunas
df_pivot = df.pivot(index='Data', columns='Produto', values='Vendas')

# Criar gráfico de linhas
plt.figure(figsize=(10, 5))
for produto in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[produto], marker='o', label=produto)

plt.title('Vendas Mensais por Produto')
plt.xlabel('Data')
plt.ylabel('Vendas (R$)')
plt.legend()  # Mostra a legenda
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
